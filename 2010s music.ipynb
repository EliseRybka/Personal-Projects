```python
from googlesearch import search
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

# Get Data

```python
df = pd.read_excel('songs.xlsx', header=None, names=['full name'])
df.head()
```

```python
def seperate_full_name(full_name, which):
    artist, song = full_name.split(' - ')
    if which == 'song':
        return song
    return artist
```

```python
df['artist'] = df.iloc[:,0].apply(seperate_full_name, which='artist')
df['song'] = df.iloc[:,0].apply(seperate_full_name, which='song')
```

```python
df['wiki url'] = 0
df['year released'] = 0
df['ya.music url'] = 0
```

```python
df.head()
```

```python
for index, row in df.iloc[115:].iterrows():
    query = row['full name']
    
    wiki_found = False
    for url in search(query,        # The query you want to run
                tld = 'com',  # The top level domain
                lang = 'en',  # The language
                num = 10,     # Number of results per page
                start = 0,    # First result to retrieve
                stop = None,  # Last result to retrieve
                pause = 2.0,  # Lapse between HTTP requests
               ):
        if 'wikipedia' in url:
            wiki_found = True
            break
            
    if not wiki_found:
        print(query)
    else:

        df.loc[index, 'wiki url'] = url



        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')

        try:
            table = soup.find(class_='infobox vevent')
            right_cells = table.find_all('td')
        except:
            table = soup.find(class_='infobox vevent haudio')
            right_cells = table.find_all('td')

        year_found = 0

        for cell in right_cells:
            entry = cell.text
            if '201' in entry:
                year_found = entry
                year = int(year_found[year_found.find('201'):year_found.find('201')+4])
                break

        if year_found == 0: raise NameError('Year Not Found')

        df.loc[index, 'year released'] = year
    
    print(index)
```

```python
for index, row in df.iterrows():
    query = row['full name'] + 'яндекс музыка'
    
    
    music_found = False
    
    for url in search(query,        # The query you want to run
                tld = 'com',  # The top level domain
                lang = 'en',  # The language
                num = 10,     # Number of results per page
                start = 0,    # First result to retrieve
                stop = None,  # Last result to retrieve
                pause = 2.0,  # Lapse between HTTP requests
               ):
        if 'music.yandex' in url:
            music_found = True
            break
    if not music_found: 
        print(row['full name'])
    else: 
        df.loc[index, 'ya.music url'] = url
```

```python
df = df.sort_values(by=['year released'], ascending=True)
df.head()
```
